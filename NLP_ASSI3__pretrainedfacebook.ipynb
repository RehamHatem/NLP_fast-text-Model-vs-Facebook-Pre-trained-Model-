```python
from huggingface_hub import hf_hub_download
import fasttext

model_path = hf_hub_download(repo_id="facebook/fasttext-en-vectors", filename="model.bin")

# Load the model using fastText
model = fasttext.load_model(model_path)
```

```python
word = "king"
word_vector = model.get_word_vector(word)
print(f"Word vector for '{word}': {word_vector}")

# the nearest words
nearest_neighbors = model.get_nearest_neighbors(word)[:10]  # Limit to the first 10 nearest words
print("Nearest words to", word, ":")
for neighbor in nearest_neighbors:
    print(neighbor[1], "-", neighbor[0])

#  furthest words
furthest_neighbors = model.get_nearest_neighbors(word, k=-10)[-10:]  # Limit to the last 10 furthest words
print("\nFurthest words from", word, ":")
for neighbor in furthest_neighbors:
    print(neighbor[1], "-", neighbor[0])
```

```python
words = ["stop", "play", "delicious", "eat", "game", "self", "phone", "grow", "happy", "sad",
               "fun", "interesting", "boring", "exciting", "tired", "sleepy", "energetic", "lazy",
               "book", "movie", "music", "dance", "sing", "art"]

for word in words:
    #  the nearest words
    nearest_neighbors = model.get_nearest_neighbors(word)[:10]  # Limit to the first 10 nearest words
    print("Nearest words to", word, ":")
    for neighbor in nearest_neighbors:
        print(neighbor[1], "-", neighbor[0])

    # the furthest words
    furthest_neighbors = model.get_nearest_neighbors(word, k=-10)[-10:]  # Limit to the last 10 furthest words
    print("\nFurthest words from", word, ":")
    for neighbor in furthest_neighbors:
        print(neighbor[1], "-", neighbor[0])
    print("\n")
```
